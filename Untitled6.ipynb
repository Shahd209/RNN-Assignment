```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
X = data.data
y = data.target

df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

print(df.shape)
print(data.feature_names)
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler

X_features = df.drop('target', axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_features)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=y, palette='Set1', s=60)
plt.title('PCA of Breast Cancer Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_pca)
df['cluster'] = clusters

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=clusters, palette='Set2', s=60)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='black', marker='X', label='Centroids')
plt.title('K-Means Clusters on PCA Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python

comparison = pd.crosstab(df['cluster'], df['target'], rownames=['Cluster'], colnames=['Actual'])
comparison.rename(columns={0: 'Malignant', 1: 'Benign'}, inplace=True)
print(comparison)
```

```python
#1. Cluster correspondence:
#From the scatter plot and comparison table, you can see that clusters mostly separate malignant vs benign.
#Some misclassifications may exist because K-Means is unsupervised.

#2. PCA impact:
#PCA reduces 30 features  2 principal components.
#Simplifies visualization & clustering while retaining most variance (~7075%).
#Helps algorithms like K-Means work better on lower-dimensional data.
```
